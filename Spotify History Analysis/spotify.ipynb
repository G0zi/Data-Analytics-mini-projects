# **Spotify History Analysis**

```python
import pandas as pd
spoti_dic = pd.read_csv('spotify_data_dictionary.csv')
spoti_his = pd.read_csv('spotify_history.csv')
```

## Which artist did they listen to the most this year? Is it the same as last year?

```python
#This year
mask = spoti_his[spoti_his['ts'] > '2024-01-01 00:00:00']
mask['Times played in 2024'] = mask['ts']
this_year = mask.groupby('artist_name').count()['Times played in 2024'].sort_values(ascending = False).index[0]
```

```python
#Last year
mask1 = spoti_his[(spoti_his['ts'] > '2023-01-01 00:00:00') & (spoti_his['ts'] < '2024-01-01 00:00:00')]
mask1['Times played in 2023'] = mask1['ts']
last_year = mask1.groupby('artist_name').count()['Times played in 2023'].sort_values(ascending = False).index[0]
```

```python
if this_year == last_year:
    print(f'This year album - {this_year} is same as last year album - {last_year}')
else:
    print(f'This year album - {this_year} is not same as last year album - {last_year}')
```

## Which songs have they played the most? How often do they skip them?

```python
most_played = spoti_his.groupby('track_name')['ms_played'].sum().sort_values(ascending = False).index[0]
print(most_played)
```

```python
mask = spoti_his[spoti_his['track_name'] ==  most_played]
mask1 = mask[mask['skipped'] == True]
print(f'No. of times {most_played} was skipped = {mask1.shape[0]}')
```

## What time of day do they typically listen to music?

```python
spoti_his['ts'] = pd.to_datetime(spoti_his['ts'])
spoti_his['time_of_the_day'] = spoti_his['ts'].dt.hour.apply(lambda x: 'Night' if x<=0 or x>=18 else
                                                             'Early morning' if x>0 or x<=6 else
                                                             'Afternoon' if x<18 or x>=12 else 
                                                             'Morning')
spoti_his
```

```python
spoti_his.groupby('time_of_the_day')['time_of_the_day'].count().sort_values(ascending = False).index[0]
```

## How often do they explore new artists versus replaying favorites?

```python
spoti_his['new_artist'] = ~spoti_his['artist_name'].duplicated()

total_plays = len(spoti_his)
new_plays = spoti_his['new_artist'].sum()
replayed = total_plays - new_plays
```

```python
print(f'Total songs played: {total_plays}')
print(f'{(new_plays/total_plays)*100:.2f}% listened to new artists')
print(f'{(replayed/total_plays)*100:.2f}% replayed favourites')
```
